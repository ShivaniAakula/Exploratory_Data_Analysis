### Visual-Analysis

```python
def univariate_num(col):
    
    print(f"The Average of {col}:",df[col].median())
    print(f"The Maximum of {col}:",df[col].max())
    print(f"The Minimum of {col}:",df[col].min())
    
    print('='*30,col,'='*30 )
    plt.figure(figsize=(5, 4))
    sns.boxplot(df[col])
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()

    plt.figure(figsize = (6,4))
    plt.hist(df[col],color = 'blue',alpha = 0.7,density = True)
    plt.xlabel(col)
    plt.ylabel('Density')
    sns.kdeplot(df[col])
    plt.show()
```

### Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['gender'], df['specialization'])

# Perform the chi-square test
chi2_stat, p_val, _, _ = chi2_contingency(contingency_table)

# Set significance level
alpha = 0.05

# Check if p-value is less than alpha
if p_val < alpha:
    print("There is a relationship between gender and specialization.")
else:
    print("There is no relationship between gender and specialization.")
```
